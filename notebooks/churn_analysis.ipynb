# Customer Churn Analysis

This notebook is a guided version of the portfolio workflow. Run `src/data_cleaning.py` first, then use this notebook for exploration, visualization, and business interpretation.

```python
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
DATA_PATH = PROJECT_ROOT / 'data' / 'processed' / 'churn_cleaned.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

## Dataset Overview

```python
df.info()
df.describe(include='all').T.head(20)
```

## Churn Rate

```python
overall_churn_rate = df['churn_flag'].mean()
print(f'Overall churn rate: {overall_churn_rate:.2%}')
```

## Churn by Contract

```python
contract_churn = df.groupby('contract')['churn_flag'].mean().sort_values(ascending=False)
contract_churn.plot(kind='bar', color='#2f6f73', figsize=(8, 4), title='Churn Rate by Contract')
plt.ylabel('Churn Rate')
plt.show()
contract_churn
```

## Churn by Tenure

```python
tenure_churn = df.groupby('tenure_group')['churn_flag'].mean()
tenure_churn.plot(kind='bar', color='#4c78a8', figsize=(8, 4), title='Churn Rate by Tenure Group')
plt.ylabel('Churn Rate')
plt.show()
tenure_churn
```

## Monthly Charges and Churn

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='churn', y='monthly_charges')
plt.title('Monthly Charges by Churn Status')
plt.show()
```
